## Data Cleaning
>Remove the duplicates in the rows(if any).

```python
#display number of duplicated
print(df.duplicated().sum())
```

>Removing Unused columns

```python
sns.heatmap(df.isna())
```

```python
# create a list of unused cloumns
delete_col=[ 'Notes', 'Status']
#delete unused columns
df= df.drop(delete_col,1)

#display data after delete unused cloumns
df.head()
```

```python
sns.heatmap(df.isna())
```

```python
#display null values
df.isnull().sum()
```

```python
df.dropna(subset=['BasePay'] ,inplace =True)

df.isnull().sum()
```

```python
#display data types of column
df.dtypes
```

```python
# change data type from string to float
df['BasePay']
```

## Exploratory Data Analysis
### a general look on the data

```python
#diplay general look
df.hist(figsize= (15,12));
```

##  1- How have salaries changed over time?

```python
salary_year = df.groupby('Year')['TotalPay'].sum()
plt.figure(figsize = (20,8))
plt.plot(salary_year)
plt.title("salary over time", fontsize = 20, fontweight = "bold", color = "darkblue")
plt.ylabel("salary",fontsize = 18)
plt.xlabel("years",fontsize = 18)
plt.show()
```

##  2- what is the name of highest paid person ?

```python
df[df['TotalPay'] == df['TotalPay'].max()]['EmployeeName']
```

## 3. what is the name of lowest paid person?

```python
df[df['TotalPay'] == df['TotalPay'].min()]['EmployeeName']
```

## 4. what is average of basepay of all employees per year?

```python
df.groupby('Year').mean()['TotalPay']
```

## 5. how many unique job title?

```python
df['JobTitle'].nunique()
```
